```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def detect_blur(image_path):
  img1=cv2.imread(image_path)
  #reading the image
  img= cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
  #applying laplacian kernel
  laplacian_var=cv2.Laplacian(img,cv2.CV_64F).var()
  #setting a threshold value
  if laplacian_var <=50:

    print("Image is as it is Blur.Please insert a clear image")
  else:
    print('Image Accepted')
 
  plt.imshow(img1)
  plt.show()
  print(laplacian_var)

  
```

```python
detect_blur('1.jpeg')
```
